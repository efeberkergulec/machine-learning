```python
### ASSIGNMENT 1 ###
# QUESTION 1
# Efe Berk ERGULEC
''' Importing Libraries. I used them during my homework. '''
import operator
import math
import numpy as np
import matplotlib.pyplot as plot
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Implementation of training number m, features x, epsilon eps and prediction y. 
m=100
x= np.random.rand(m,1)
eps = np.random.randn(m,1)
y = np.sin(2 * np.pi * x) + eps
plot.plot(x,y,'b.')
```

```python
# d = 0. Zeroth order Polynomial Model. It hasn't got any order so it will be a line without any slope.
polynomial_features_zero = PolynomialFeatures(degree=0)
x_poly_zero = polynomial_features_zero.fit_transform(x)

model_zero = LinearRegression()
model_zero.fit(x_poly_zero, y)
y_poly_pred_zero = model_zero.predict(x_poly_zero)

plot.scatter(x, y, s=10, c='b')
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred_zero), key=sort_axis)
x_zero, y_poly_pred_zero = zip(*sorted_zip)
plot.plot(x_zero, y_poly_pred_zero, color='m')
plot.show()
```

```python
# d = 1. First order Polynomial Model which is Linear Regression.
x_one = x[:]
model = LinearRegression()
model.fit(x_one, y)
y_pred_one = model.predict(x_one)

plt.plot(x_one,y,'b.')
plt.plot(x_one, y_pred_one, color='r')
plt.show()
```

```python
# d = 3. Third order Polynomial Model.
polynomial_features_three = PolynomialFeatures(degree=3)
x_poly_three = polynomial_features_three.fit_transform(x)

model_three = LinearRegression()
model_three.fit(x_poly_three, y)
y_poly_pred_three = model_three.predict(x_poly_three)

plot.scatter(x, y, s=10, c='b')
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred_three), key=sort_axis)
x_three, y_poly_pred_three = zip(*sorted_zip)
plot.plot(x_three, y_poly_pred_three, color='y')
plot.show()
```

```python
# d = 9. Ninth order Polynomial Model.
polynomial_features_nine = PolynomialFeatures(degree=9)
x_poly_nine = polynomial_features_nine.fit_transform(x)

model_nine = LinearRegression()
model_nine.fit(x_poly_nine, y)
y_poly_pred_nine = model_nine.predict(x_poly_nine)

plot.scatter(x, y, s=10, c='b')
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred_nine), key=sort_axis)
x_nine, y_poly_pred_nine = zip(*sorted_zip)
plot.plot(x_nine, y_poly_pred_nine, color='c')
plot.show()
```

```python
# Plot graph of all experiments
plot.scatter(x, y, s=10, c='b')
plot.plot(x_zero, y_poly_pred_zero, color='m', label='d = 0')
plt.plot(x_one, y_pred_one, color='r', label='d = 1')
plot.plot(x_three, y_poly_pred_three, color='y', label='d = 3')
plot.plot(x_nine, y_poly_pred_nine, color='c', label='d = 9')
plot.title('Experiment #1')
plot.legend()
plot.show()
```
